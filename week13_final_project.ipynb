```python
        
def readInFile(rel_path):
    text_file = open(rel_path, "r", encoding="utf-8")
    content = text_file.read()
    words = content.split('\n')
    text_file.close()
    return words

def translit(words, lang):
    epi = epitran.Epitran(lang)
    ipa = []
    for word in words:
        ipa.append(epi.transliterate(word))
    return ipa

def buildNPath(abbrev):
    return "C:/Users/Clair/machine_learning_sp20/Final_Project/sentiment_lexicons/negative_words_" + abbrev + ".txt"

def buildPPath(abbrev):
    return "C:/Users/Clair/machine_learning_sp20/Final_Project/sentiment_lexicons/positive_words_" + abbrev + ".txt"


```

```python
n_ipas = []
p_ipas = []

length = 0;
i = 0

for lang in langs:

    n_words = readInFile(buildNPath(abbrevs[i]))
    p_words = readInFile(buildPPath(abbrevs[i]))

    n_ipa = translit(n_words, lang)
    p_ipa = translit(p_words, lang)
       
    n_ipas.extend(n_ipa)
    p_ipas.extend(p_ipa)
    print("Lang =", abbrevs[i], "completed")

    i += 1;
```

```python
comb_chars = []
    
for i in range(42752, 42783):
    comb_chars.append(chr(i))
    
for i in range(688, 879):
    comb_chars.append(chr(i))
```

```python
up_sym = '\u0361'
colon_sym = '\u02D0'

def is_combo_char(char):
    return (char in comb_chars)

def has_combo_char(word):
    for letter in word:
        if is_combo_char(letter):
            return True;
    return False;

def split(word): 
    return [char for char in word] 
```

```python
def get_up_token(word, index):
    return word[index] + word[index + 1] + word[index + 2]

def get_combo_token(word, index):
    return word[index] + word[index + 1]

def get_list_with_up_symb(word):
    new_word_lst = []
    i = 0
    while (i < len(word)):
        if (i + 1 < len(word) and word[i + 1] == up_sym):
            new_word_lst.append(get_up_token(word, i))
            i += 3
        else:
            new_word_lst.append(word[i])
            i += 1
            
    return new_word_lst

def get_list_with_combo(word):
    new_word_lst = []
    i = 0
    while (i < len(word)):
        if (i + 1 < len(word) and is_combo_char(word[i + 1])):
            new_word_lst.append(get_combo_token(word, i))
            i += 2
        else:
            new_word_lst.append(word[i])
            i += 1
            
    return new_word_lst
```

```python
word = n_ipas[3] + n_ipas[67] + n_ipas[3]
word_id = 44
word_as_list = []

us_count = word.count(up_sym)
has_combo = has_combo_char(word)
word_as_list = split(word)        

if (us_count > 0 or has_combo):
    if (us_count > 0):
        word_as_list = get_list_with_up_symb(word_as_list)
    if (has_combo):
        word_as_list = get_list_with_combo(word_as_list)
```

```python
n_list = []
p_list = []

wordId = 0

for i in range(len(n_ipas)):
    word = n_ipas[i]

    us_count = word.count(up_sym)
    has_combo = has_combo_char(word)
    word_as_list = split(word)        

    if (us_count > 0 or has_combo):
        if (us_count > 0):
            word_as_list = get_list_with_up_symb(word_as_list)
        if (has_combo):
            word_as_list = get_list_with_combo(word_as_list)

    n_list.append([wordId] + word_as_list)
    wordId += 1
    symbCount = 0

wordId = 0
for i in range(len(p_ipas)):
    word = p_ipas[i]

    us_count = word.count(up_sym)
    has_combo = has_combo_char(word)
    word_as_list = split(word)        

    if (us_count > 0 or has_combo):
        if (us_count > 0):
            word_as_list = get_list_with_up_symb(word_as_list)
        if (has_combo):
            word_as_list = get_list_with_combo(word_as_list)

    p_list.append([wordId] + word_as_list)
    wordId += 1
    symbCount = 0
```
