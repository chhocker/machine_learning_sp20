```python
import numpy as np
import pandas as pd
import keras

from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
#################################################################
# DATASET #1
# Breast Cancer Tumor Analysis
# -Classification
#################################################################
cancer_columns = ["id_num", "diagnosis", "radius1", "texture1", "perimeter1", "area1", "smoothness1", 
                  "compactness1", "concavity1", "concave_pts1", "symmetry1", "fract_dimen1",
                 "radius2", "texture2", "perimeter2", "area2", "smoothness2", 
                  "compactness2", "concavity2", "concave_pts2", "symmetry2", "fract_dimen2",
                 "radius3", "texture3", "perimeter3", "area3", "smoothness3", 
                  "compactness3", "concavity3", "concave_pts3", "symmetry3", "fract_dimen3"]
cancer_data = pd.read_csv("wdbc.data", delimiter=",", names=cancer_columns, header=None)
```

```python
cancer_data["is_malig"] = cancer_data.diagnosis.map({"M": 1, "B": 0})

cancer_targets = cancer_data["is_malig"]
cancer_data = cancer_data.drop(columns=["id_num", "diagnosis", "is_malig"])

cancer_data = cancer_data[:].values
```

```python
train_data, test_data, train_targets, test_targets = train_test_split(cancer_data, cancer_targets, 
                                                                      test_size = 0.3, shuffle = True)
```

```python
classifier = Sequential()
```

```python
#Input layer with 5 inputs neurons
classifier.add(Dense(activation="relu", input_dim=30, units=3, kernel_initializer="uniform"))
#Hidden layer
classifier.add(Dense(activation="relu", units=2, kernel_initializer="uniform"))

#output layer with 1 output neuron which will predict 1 or 0
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

classifier.fit(train_data, train_targets, batch_size = 10, epochs = 100)
```

```python
#getting predictions of test data
prediction = classifier.predict(test_data).tolist()
# list to series
se = pd.Series(prediction)
# creating new column of predictions in data_check dataframe
final = pd.DataFrame()
final['check'] = se
final['check'] = final['check'].str.get(0)

series = []
for val in final['check']:
    if val >= 0.5:
        series.append(1)
    else:
        series.append(0)
        
match = 0
nomatch = 0
i = 0
for true in test_targets:
    if true == series[i]:
        match = match + 1
    else:
        nomatch = nomatch + 1
    i += 1

print(match, nomatch)
print("Percent correctly classified:", 100 * round(match / 171, 2))
```

```python
#################################################################
# DATASET #2
# Forest Fires
# -Regression
# -Label Encoding
#################################################################
```

```python
# READ IN DATA
fire_columns = ["x_coord", "y_coord", "month", "day", "ffmc", "dmc", "dc", 
                  "isi", "temp", "rel_humid", "wind", "rain",
                 "area"]
fire_data = pd.read_csv("forestfires.csv", names=fire_columns, header=None)
```

```python
# ENCODE STRING DATA
fire_data["month_num"] = fire_data.month.map({"jan": 1, "feb": 2, "mar": 3, "apr": 4, "may": 5, "jun": 6, 
                                              "jul": 7, "aug": 8, "sep": 9, "oct": 10, "nov": 11, "dec": 12})

fire_data["week_day_num"] = fire_data.day.map({"sun": 1, "mon": 2, "tue": 3, "wed": 4, "thu": 5, "fri": 6, "sat":7})

fire_data = fire_data.drop(columns=["month", "day"])
```

```python
fire_targets = fire_data["area"]
fire_data = fire_data.drop(columns=["area"])
```

```python
train_data, test_data, train_targets, test_targets = train_test_split(fire_data, fire_targets, 
                                                                      test_size = 0.3, shuffle = True)
classifier = Sequential()
```

```python
#Input layer with 5 inputs neurons
classifier.add(Dense(activation="relu", input_dim=12, units=3, kernel_initializer="uniform"))
#Hidden layer
classifier.add(Dense(activation="relu", units=2, kernel_initializer="uniform"))

#output layer with 1 output neuron which will predict 1 or 0
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

classifier.fit(train_data, train_targets, batch_size = 10, epochs = 100)
```

```python
#getting predictions of test data
prediction = classifier.predict(test_data).tolist()
# list to series
se = pd.Series(prediction)
# creating new column of predictions in data_check dataframe
final = pd.DataFrame()
final['check'] = se
final['check'] = final['check'].str.get(0)
```
